```python
print("=== Secure Notes App ===")

choice = input("Choose add / view: ").strip().lower()

if choice == "add":
    note = input("Enter note: ").strip()

    # 1. Reject empty notes
    if note == "":
        print("Error: Note cannot be empty.")
    else:
        encrypted = ""

        # ENCRYPTION:
        # Each character is converted to its ASCII value,
        # shifted forward by 1, then converted back to a character.
        for ch in note.lower():
            encrypted = encrypted + chr(ord(ch) + 1)

        file = open("secure_notes.txt", "a")
        file.write(encrypted + "\n")
        file.close()

        print("Note saved securely")

elif choice == "view":
    file = open("secure_notes.txt", "r")

    count = 0
    total_chars = 0

    for line in file:
        decrypted = ""

        # DECRYPTION:
        # Each character is converted to its ASCII value,
        # shifted backward by 1, then converted back to a character.
        for ch in line.strip():
            decrypted = decrypted + chr(ord(ch) - 1)

        # 2. Convert output to title case
        decrypted = decrypted.title()

        print("- " + decrypted)

        # 3. Count total characters across all notes
        total_chars = total_chars + len(decrypted)
        count = count + 1

    file.close()

    print("Total notes:", count)
    print("Total characters:", total_chars)

else:
    print("Invalid option.")
```

# Task
Develop a Python-based menu-driven feedback application with the following features: add new feedback, view all existing feedback, analyze feedback (starting with a simple count), search feedback by keyword, and gracefully exit the application.

## Initialize Application

### Subtask:
Set up the main application loop and display the menu options to the user. This will use a `while` loop and prompt the user to choose an option (add, view, analyze, search, exit).


**Reasoning**:
I need to initialize the application loop by starting a `while True:` loop, displaying the menu options, and capturing user input, as per the instructions.



```python
print("=== Secure Notes App ===")

while True:
    print("\nMenu:")
    print("1. Add Feedback")
    print("2. View Feedback")
    print("3. Analyze Feedback")
    print("4. Search Feedback")
    print("5. Exit")

    choice = input("Choose an option: ").strip().lower()

    if choice == "1" or choice == "add feedback":
        feedback = input("Enter your feedback: ").strip()

        # 1. Reject empty feedback
        if feedback == "":
            print("Error: Feedback cannot be empty.")
        else:
            encrypted = ""

            # ENCRYPTION:
            # Each character is converted to its ASCII value,
            # shifted forward by 1, then converted back to a character.
            for ch in feedback.lower():
                encrypted = encrypted + chr(ord(ch) + 1)

            try:
                file = open("feedback.txt", "a")
                file.write(encrypted + "\n")
                file.close()
                print("Feedback saved securely.")
            except Exception as e:
                print(f"Error saving feedback: {e}")

    elif choice == "2" or choice == "view feedback":
        try:
            file = open("feedback.txt", "r")

            count = 0
            total_chars = 0

            for line in file:
                decrypted = ""

                # DECRYPTION:
                # Each character is converted to its ASCII value,
                # shifted backward by 1, then converted back to a character.
                for ch in line.strip():
                    decrypted = decrypted + chr(ord(ch) - 1)

                # 2. Convert output to title case
                decrypted = decrypted.title()

                print("- " + decrypted)

                # 3. Count total characters across all feedback
                total_chars = total_chars + len(decrypted)
                count = count + 1

            file.close()

            print("Total feedback entries:", count)
            print("Total characters:", total_chars)
        except FileNotFoundError:
            print("No feedback entries found yet.")
        except Exception as e:
            print(f"Error viewing feedback: {e}")

    elif choice == "3" or choice == "analyze feedback":
        print("Analyze Feedback functionality not yet implemented.")
    elif choice == "4" or choice == "search feedback":
        print("Search Feedback functionality not yet implemented.")
    elif choice == "5" or choice == "exit":
        print("Exiting Secure Notes App. Goodbye!")
        break
    else:
        print("Invalid option. Please choose a valid option.")
```
